## ETS - Error, Trend and Seasonality

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("airline-passengers.csv", index_col=0, parse_dates=True)
df.head(2)
```

```python
df = df.dropna()
```

```python
df.plot(figsize=(12,5))
```

```python
### Data trend is either increasing or decreasing --> use multiplicative model

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df['Passengers'], model='multiplicative')
```

```python
## to increase the plot size use pylab
from pylab import rcParams
rcParams['figure.figsize'] = 12,5

result.plot();
```

```python
## to plot individual result

result.trend.plot(figsize=(12,5))
```
